```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
import warnings
warnings.filterwarnings("ignore")
```

### Подготовим текст Войны и Мира

```python
all_text = ''
for i in range(1,5):
    with open(f'war_and_peace/tom_{i}.txt', 'r', encoding='utf-8') as f:
        text = f.read().lower()
        all_text += re.sub('[^а-я]', '', text)
```

### Посчитаем матрицу вероятностей

```python
russian = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'

POS = {l: i for i, l in enumerate(russian)}

probabilities = np.zeros((len(russian), len(russian)))
```

```python
for i, j in zip(all_text[:-1], all_text[1:]):
    probabilities[POS[i], POS[j]] +=1
```

```python
probabilities = probabilities/probabilities.sum(axis=1)[:, None]
```

```python
df = pd.DataFrame(probabilities, index=(i for i in russian), columns=(i for i in russian))
df.head(10)
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(df)
plt.show()
```

### Вероятность встретить строку из 4х символов, начиная с "П", заканчивая "Р"

```python
def get_4_character_word(start_letter, end_letter):
    
    start = start_letter
    end = end_letter
    first_value = 1.0/33
    proba = 0.
    
    high_proba = 0.
    word = ''
    
    for i in russian:
        for j in russian: 
            proba = first_value* probabilities[POS[start], POS[i]] * probabilities[POS[i], POS[j]] * probabilities[POS[j], POS[end]]
            if proba > high_proba:
                high_proba = proba
                word = start + i + j + end
            if proba > 0:
                print(''.join([start, i, j, end]), proba)

    print('--------')
    print(word, high_proba)
```

```python
get_4_character_word('п', 'р')
```

#### Получается, для рассчета вероятности любого слова сложность данного алгоритма будет 33^(len(word) - 2), если считать для Русского языка

### Генерация слова

```python
def get_random_word(start_letter, length):
    s = [start_letter]
    for i in range(length):
        index = POS[s[i]]
        next_letter = np.random.choice(list(russian), p=probabilities[index])
        s.append(next_letter)
    print(''.join(s))
```

```python
get_random_word('б', 8)
```

### Оценка ряда с помощью HMM

```python
from hmmlearn.hmm import GaussianHMM
```

```python
df = pd.read_csv('Series/weekly-closings-of-the-dowjones-.csv')
series = df['Close']
```

```python
plt.figure(figsize=(14,6))
plt.plot(series)
```

#### Продифференцируем, уберем тренд

```python
series = np.diff(series, 1)
```

```python
plt.figure(figsize=(14,6))
plt.plot(series)
```

```python
values = series.reshape(-1,1)
```

### Обучим Марковскую модель 

```python
for n in range(1, 14):
    model = GaussianHMM(n_components=n,
                        covariance_type="diag",
                        n_iter=1000)
    model.fit(values)
    print (n, model.score(values))
```

```python
model = GaussianHMM(n_components=8, covariance_type="diag", n_iter=1000)
model.fit(values)
print(model.score(values))
```

```python
labels = model.predict(values)
means = np.zeros_like(values)
    
for i in range(len(labels)):
    means[i] = model.means_[labels[i]]
    
plt.figure(figsize=(12, 6))
plt.plot(values)
plt.plot(means, linewidth=3)
```

### Функция для предсказания на основе случайных Марковских процессов

```python
def get_predict(model, values, n_pred):
    means = []
    
    s = list(model.predict(values))
    components = [i for i in range(model.n_components)]
    
    # Предсказываем новые значения на основе случайных Марковских процессов
    for i in range(n_pred):
        index = s[-1]
        next_label = np.random.choice(components, p=model.transmat_[index])
        s.append(next_label)
        
    for i in range(len(s)):
        means.append(model.means_[s[i]])
    
    plt.figure(figsize=(12, 6))
    plt.plot(values)
    plt.plot(means, linewidth=3)
    
    return means
```

```python
m = get_predict(model, values, 10)
```
